```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20, 15)
import yt
import seaborn as sns
import h5py
```

```python
fields = {
    'density': ('gas', 'density'),
    'pressure': ('gas', 'pressure')
}
```

```python
convolution_file = h5py.File('convolution_2d_RD0000.hdf5')
density_data = np.array(convolution_file['density'][:])*10e40
pressure_data = np.array(convolution_file['pressure'][:])*10e40
density_data = np.array(density_data)
```

```python
density_image = yt.ImageArray(density_data, info={'field': fields['density']})
pressure_image = yt.ImageArray(pressure_data, info={'field': fields['pressure'], 'units':'cgs'})
```

```python
plt.imshow(np.log10(density_data))
```

```python
plt.imshow(np.log10(pressure_image))
```
